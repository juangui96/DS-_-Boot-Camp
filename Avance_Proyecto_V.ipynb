## GRUPO QUERY 1: Cual es la tasa de ventas por mes y cuales son los meses que tienen una mayor venta por dia?
Segun la query que se muestra a continuacion, se puede ver que el mes de septiembre, octubre, noviembre y diciembre son los que mejor comportamiento presenta en cuanto a ventas/dia. Asi mismo, el año 2017 y 2018 son los que tienen un valor mas alto en cuanto a esta tasa, esto se puede apreciar en la segunda query.

```python
query_tasaventas = """
WITH Ventas_mes as
(SELECT
    AÑO, Mes, CAST(sales2 as FLOAT) AS Ventas_miles,
    CASE
      WHEN Mes in (1,3,5,7,8,10,12) THEN 31
      WHEN Mes in (4,6,9,11) THEN 30
      ELSE CAST('28' as FLOAT)
    END AS Dias_mes
FROM ss1)
SELECT AÑO,Mes, sum(Ventas_miles), Dias_mes, (sum(Ventas_miles)/Dias_mes) as Tasa
FROM Ventas_mes
GROUP BY AÑO,Mes
ORDER BY Tasa Desc
"""
tasa_ventas = runQuery(query_tasaventas)
tasa_ventas
```

```python
query_ventasregion1 = """
WITH vta_clt AS (
  SELECT AÑO, "Customer Name", State, City, SUM(Sales2) AS total_sales
  FROM ss1
  GROUP BY AÑO, "Customer Name", State, City
),
prom_vta AS (
  SELECT AÑO, AVG(Sales2) AS avg_sales
  FROM ss1
  GROUP BY AÑO
),
join_vtas AS (
  SELECT v.*, p.avg_sales
  FROM vta_clt v
  LEFT JOIN prom_vta p ON p.AÑO = v.AÑO
),
vtas_final AS (
SELECT *, CASE
  WHEN total_sales >= avg_sales THEN "Encima Prom"
  ELSE "Debajo prom"
  END AS Comparacion
FROM join_vtas
WHERE Comparacion = 'Encima Prom')
SELECT "Customer name", AÑO,  state, city, count(Comparacion)
FROM vtas_final
GROUP BY "Customer Name", AÑO, comparacion
ORDER BY count(Comparacion) DESC
"""
ventas_region = runQuery(query_ventasregion1)
ventas_region
```

```python
query_varcliente = """
WITH vtas_2015 AS (
  SELECT  "Customer Name" as CN, SUM(Sales2) AS total_sales_2015
  FROM ss1
  WHERE AÑO = '2015'
  GROUP BY "Customer Name"
),
vtas_2016 AS (
  SELECT  "Customer Name" as CN, SUM(Sales2) AS total_sales_2016
  FROM ss1
  WHERE AÑO = '2016'
  GROUP BY "Customer Name"
),
vtas_2017 AS (
  SELECT "Customer Name" as CN, SUM(Sales2) AS total_sales_2017
  FROM ss1
  WHERE AÑO = '2017'
  GROUP BY "Customer Name"
),
vtas_2018 AS (
  SELECT "Customer Name" as CN, SUM(Sales2) AS total_sales_2018
  FROM ss1
  WHERE AÑO = '2018'
  GROUP BY "Customer Name"
),
vtas_final AS (
SELECT vtas_2015.CN, vtas_2015.total_sales_2015, vtas_2016.total_sales_2016, vtas_2017.total_sales_2017, vtas_2018.total_sales_2018
FROM vtas_2015
JOIN vtas_2016 ON vtas_2016.CN = vtas_2015.CN
JOIN vtas_2017 ON vtas_2017.CN = vtas_2015.CN
JOIN vtas_2018 ON vtas_2018.CN = vtas_2015.CN)

SELECT CN,
  ((total_sales_2016-total_sales_2015)/total_sales_2015) AS VARIACION_AÑO1,
  ((total_sales_2017-total_sales_2016)/total_sales_2016) AS VARIACION_AÑO2,
  ((total_sales_2018-total_sales_2017)/total_sales_2017) AS VARIACION_AÑO3
FROM vtas_final
WHERE
((total_sales_2016-total_sales_2015)/total_sales_2015)>0 AND
((total_sales_2017-total_sales_2016)/total_sales_2016)>0 AND
((total_sales_2018-total_sales_2017)/total_sales_2017) >0
ORDER BY VARIACION_AÑO1, VARIACION_AÑO2, VARIACION_AÑO3
"""
varcliente = runQuery(query_varcliente)
varcliente
```
